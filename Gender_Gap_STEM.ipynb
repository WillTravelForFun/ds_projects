# Gender Gap in Higher Education

This project was done as part of [Dataquest.io](https://www.dataquest.io/) Data Science curriculum. The dataset used here can be downloaded from [Dr. Randal S. Olson's website](http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

women_degrees = pd.read_csv('C:\\Users\\Katya\\Documents\\Code\\percent-bachelors-degrees-women-usa.csv')
women_degrees.shape
```

The dataframe consists of 42 rows and 18 columns and contains data about college graduates from 1970 to 2011. Specifically, it shows what percentage of degree recipients across the college majors were women. For instance, here's how the percentage of women graduating with a degree in Biology has changed over time.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax = plt.plot(women_degrees["Year"], women_degrees["Biology"], c="orange")
plt.tick_params(bottom="off", top="off", left="off", right="off")
plt.title("Biology Degrees Conferred to Women, 1970 - 2011")
plt.ylabel("Percentage of Degrees")
plt.show()
```

The percentage of women graduating with a degree in Biology has doubled since 1970, but the upward trend becomes even more obvious when compared to the percentage of men graduating with the same Bachelor's degree. This data can be inferred from the dataset, since women's representation in each field is listed as a percentage.

```python
men_degrees_biology = 100 - women_degrees["Biology"]
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(women_degrees["Year"], women_degrees["Biology"], c="orange", label="Women")
plt.plot(women_degrees["Year"], men_degrees_biology, c="blue", label="Men")
plt.legend(loc="upper right")
plt.tick_params(bottom="off", top="off", left="off", right="off")
plt.title("Percentage of Biology Degrees Awarded By Gender")
plt.show()
```

In the field of Biology, women have been outnumbering men in every graduating class since about 1987. In the space below the  line graph was recreated for each major listed in the dataset.

```python
col_names = women_degrees.columns.tolist()
```

```python
major_categories = col_names[1:]
print(len(major_categories))
```

The variable col_names is a list of column headings. The first item was removed because it represents the year for which the data was gathered, and the remainder of the list was saved as a new variable, "major_categories." This list, "major_categories", was then passed in the loop below to pull data for each major sequentially. Chart spines and ticks were also removed to focus readers' attention on the trends themselves. 

```python
fig = plt.figure(figsize=(16,48))

for sp in range(0,17):
    ax = fig.add_subplot(9,2,sp+1)
    ax.plot(women_degrees["Year"], women_degrees[major_categories[sp]], c="orange", label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[major_categories[sp]], c="blue", label="Men")
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.set_title(major_categories[sp])
plt.legend(loc="upper right")
plt.show()
```

Computer Science and Engineering seem to be the two areas where women are still underrepresented. Interestingly, Public Administration, Psychology, English, Education, Foreign Languages, and Health Professions have a much larger proportion of women than men.
