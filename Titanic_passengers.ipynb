# Titanic Passengers

This project was done as part of [Dataquest.io](https://www.dataquest.io/) Data Science curriculum. The dataset used here is the training set from [Kaggle](https://www.kaggle.com/c/titanic/data?train.csv).

I started by importing the necessary libraries, and reading the dataset file into a dataframe. Then I selected the columns I wanted to keep (there were several columns that I did not intend to use in the analysis, such as passenger names), and removed rows with missing values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv("C:\\Users\\Katya\\Documents\\Code\\train.csv")
titanic = titanic[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
titanic = titanic.dropna()

#dataset preview
titanic.head()
```

I set out to create kernel density plots using Seaborn. First I subsetted the data into groups that I wanted to compare (using the FacetGrid class). There were 3 characteristics I was interested in: passengers' class, gender, and whether or not they survived. All six plots have "Age" graphed on the x-axis. 

```python
g = sns.FacetGrid(titanic, col="Pclass", row="Survived", hue="Sex", size=4)
g.map(sns.kdeplot, "Age", shade=True)
g.add_legend()
sns.despine(left=True, bottom=True)
plt.show()
```
